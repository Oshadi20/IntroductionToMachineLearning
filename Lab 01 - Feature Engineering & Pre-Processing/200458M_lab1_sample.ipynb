### Handling missing values

```python
# plot the missing data heatmap
# train data
plt.figure(figsize=(10,8))

cols = df_train.columns[:15]
colours = ['#000099', '#ffff00'] # specify colours: yellow - missing, blue - not missing
sns.heatmap(df_train[cols].isna(), cmap=sns.color_palette(colours))
```

```python
# test data
plt.figure(figsize=(10,8))

cols = df_test.columns[:]
colours = ['#000099', '#ffff00'] # specify colours: yellow - missing, blue - not missing
sns.heatmap(df_test[cols].isna(), cmap=sns.color_palette(colours))
```

```python
#impute the missing values with their respective medians
df_train = df_train.fillna(df_train.mean())
df_test = df_test.fillna(df_test.mean())
```

### Correlation between data

```python
# create correlation matrix
corr_matrix = df_train.corr().abs()
corr_matrix
```

```python
# select upper traingle of correlation matrix
upper_corr = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
upper_corr
```

```python
# Find index of columns with correlation greater than 0.95
drop_cols = [column for column in upper_corr.columns if any(upper_corr[column] > 0.95)]
```

```python
# x10 is highly correlated with the target
# therefore drop the x11 column
df_train = df_train.drop('x11',axis = 1)
df_test = df_test.drop('x11',axis = 1)
df_train.head(5)
```

### Figure out and handle outliers

```python
from scipy import stats
# calculate Z-scores for each value in the dataset
z_scores = stats.zscore(df_train)
```

```python
# define a thershold for the Z-score values to identify outliers
threshold = 3.4 # values that are more than 3.4 standard deviations away from the mean

# identify the rows in the dataset that contain outliers
outliers = df_train[(z_scores > threshold) | (z_scores < -threshold)]
```

```python
# handle the outliers
# cap the outliers
for col in df_train.columns:
    max_val = df_train[col].max()
    min_val = df_train[col].min()
    df_train[col] = df_train[col].clip(lower=min_val, upper=max_val)
```

### Splitting the data

```python
# split the train data into x and y
X_train = df_train.drop(columns=['y'])
y_train = df_train['y']
```

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
test_scalar = StandardScaler()
X_train_scaled = test_scalar.fit_transform(X_train)
X_test_scaled = test_scalar.fit_transform(X_test)
```

```python
# convert receiving numpy array into pandas dataframe
X_train = pd.DataFrame(X_train_scaled, index=X_train.index, columns=X_train.columns)
X_train.head()
```

```python
X_test = pd.DataFrame(X_test_scaled, index=X_test.index, columns=X_test.columns)
X_test.head()
```

```python
X_train.describe().round(2)
```

```python
X_test.describe().round(2)
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif
```

```python
# feature extraction
mic_best = SelectKBest(score_func=mutual_info_classif, k = 10) #select best 10 features
```

```python
# fit the selector on the training data and transform the data
X_train_selected = mic_best.fit_transform(X_train, y_train)
X_train_selected.shape
```

```python
# get the indices of the selected features
selected_features_indices = mic_best.get_support(indices=True)
```

```python
# selected columns
selected_columns = X_train.iloc[:,selected_features_indices].columns.tolist()
selected_columns
```

```python
# get only the selected features
X_train = X_train[selected_columns]
X_test = X_test[selected_columns]
```

```python
X_train.shape
```

```python
X_test.shape
```

### Train and predict

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# instantiate learning model (k = 35)
knn_classifier = KNeighborsClassifier(n_neighbors=35)

# fitting the model
knn_classifier.fit(X_train, y_train)

# predicting the Test set results
y_pred = knn_classifier.predict(X_test)
```

### Evaluating the predictions

```python
# calculating model accuracy
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of the kNN classifier is equal ' + str(round(accuracy, 2)) + ' %.')
```
